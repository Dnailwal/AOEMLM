```python
mn=0
mx=40

ds91=ds[(ds['YEAR']==1991) & (ds['SZA']>=mn)& (ds['SZA']<=mx)]
ds92=ds[(ds['YEAR']==1992)& (ds['SZA']>=mn)& (ds['SZA']<=mx)]
ds93=ds[(ds['YEAR']==1993)& (ds['SZA']>=mn)& (ds['SZA']<=mx)]
ds94=ds[(ds['YEAR']==1994)& (ds['SZA']>=mn)& (ds['SZA']<=mx)]
ds95=ds[(ds['YEAR']==1995)& (ds['SZA']>=mn)& (ds['SZA']<=mx)]
ds96=ds[(ds['YEAR']==1996)& (ds['SZA']>=mn)& (ds['SZA']<=mx)]
ds97=ds[(ds['YEAR']==1997)& (ds['SZA']>=mn)& (ds['SZA']<=mx)]
#ds98=ds[(ds['YEAR']==1998)]
ds99=ds[(ds['YEAR']==1999)& (ds['SZA']>=mn)& (ds['SZA']<=mx)]
ds20=ds[(ds['YEAR']==2000)& (ds['SZA']>=mn)& (ds['SZA']<=mx)]

year=ds['YEAR'].unique()
vr=[ds91['VER'].mean(), ds92['VER'].mean(), ds93['VER'].mean(), ds94['VER'].mean(), ds95['VER'].mean(), 
    ds96['VER'].mean(), ds97['VER'].mean(), ds99['VER'].mean(), ds20['VER'].mean()]

f107=[208, 150.1, 109.6, 85.6, 77.2, 72, 80.9, 153.9, 180]

plt.figure(figsize=(4.5,3.5))
plt.plot(year,f107, c='r',marker='*',label='F10.7')
plt.xlabel('Year',fontsize=12)
plt.ylabel('F10.7 (SFU)',fontsize=11)
plt.legend(bbox_to_anchor=(0.1, 0.95))
plt.text(1995.5,205,'(a)',fontsize=12)

plt.twinx()
plt.plot(year,vr, c='g',marker='*',label='[O] VER')
plt.ylabel('VER (Photon/$cm^3$/sec)',fontsize=11)
plt.legend(bbox_to_anchor=(0.95, 0.95))

#plt.savefig('F107_VER_annual_SZA_0_to_40.png',dpi=300, bbox_inches='tight')
#plt.savefig('F107_VER_annual_SZA_0_to_40.eps',dpi=300, bbox_inches='tight')
```
